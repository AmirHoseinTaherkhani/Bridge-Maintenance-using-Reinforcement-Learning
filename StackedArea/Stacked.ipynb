```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font',family='Arial')
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14)
```

```python
data = pd.read_csv('..//AlphaEffect//Simulations//Alpha0_simulated.csv')[:20]
```

```python
years = np.arange(1, 21)
action_cost = data['action_cost'].values
failure_cost = data['failure_risk'].values
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.set_xlim(1,20)
ax.set_ylim(0, 30000)
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_xlabel('Year', fontname = 'Arial', fontsize = 16)
ax.set_ylabel('Cost($)', fontname = 'Arial', fontsize = 16)


pal = ['steelblue', 'silver']
ax.stackplot(np.arange(1,21),action_cost, failure_cost, colors=pal,
             labels=['Action cost= $60k','Expected cost of failure = $72k'], alpha = .81)

plt.plot(np.arange(1,21), action_cost + failure_cost,lw = 1, linestyle = 'dashed', color = 'black',
         label = 'Total cost = $132k', alpha = .80)

plt.grid()


plt.legend(fontsize = 14, loc= 'upper left')
plt.tight_layout()
plt.savefig('Alpha0.jpg', dpi = 300)
```
