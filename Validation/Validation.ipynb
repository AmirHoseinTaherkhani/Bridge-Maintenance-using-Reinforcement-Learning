```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('data.xlsx')
```

```python
plt.figure(figsize = (7, 4))
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14)
plt.rc('font', **{'family' : 'Arial','weight' : 'normal'})

plt.xlabel('Year', fontname="Arial", fontsize=16)
plt.ylabel('Cumulative Cost ($1k)', fontname="Arial", fontsize=16)
plt.xlim(1,20)
# plt.ylim(0,250)

plt.plot(np.arange(1, 21), data['Alpha0']/1000, lw=3, color = 'tab:green',\
         label = '\u03B1 = 0, episode average cost = $6.6K')


plt.plot(np.arange(1, 21), data['Alpha025']/1000, color = 'tab:blue',\
         label = '\u03B1 = 0.25, episode average cost = $11.3K')
        

plt.plot(np.arange(1, 21), data['Alpha05']/1000, color = 'tab:purple',\
         label = '\u03B1 = 0.50, episode average cost = $9.4K')


plt.plot(np.arange(1, 21), data['Alpha075']/1000, color = 'tab:orange',alpha = .5,\
         label = '\u03B1 = 0.75, episode average cost = $13.7K')


plt.plot(np.arange(1, 21), data['Alpha1']/1000, '--', color = 'tab:gray',\
         label = '\u03B1 = 1, episode average cost = $8.2K')


# plt.title('yearly cost')
plt.legend()
plt.tight_layout()
plt.savefig('Validation.jpg', dpi = 300)
```
